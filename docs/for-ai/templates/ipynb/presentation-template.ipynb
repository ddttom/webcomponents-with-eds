**About this template:** This template uses pure markdown for auto-wrapping in notebook mode.

**Auto-wrapping:** When displayed in **notebook mode** (`| IPynb Viewer (notebook) |`), cells are automatically wrapped with beautiful styling:
- First cell with `#` heading ‚Üí Hero cell
- Early cells with `##` ‚Üí Intro cards (thick border)
- Short cells (‚â§3 lines) ‚Üí Transition cards
- Other cells ‚Üí Content cards (thin border)

**For other display modes:** If you need compatibility with paged, autorun, or basic modes, use manual HTML styling instead (see presentation-template-styled.ipynb).

**Hybrid approach:** Mix pure markdown with custom HTML for specific cells needing special styling.

**Use this for:**
- Fast authoring (90% less code)
- Notebook mode display
- Non-interactive presentations
- Quick demos and showcases
